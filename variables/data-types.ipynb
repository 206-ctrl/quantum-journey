```python
name = 'Lamborghini'    # string because it is a sequence of characters enclosed in single or double quotation marks
price = 540,000    # integer because it is a whole number without decimals
height_in_inches = 47.5    # float because it is a number with a decimal point
over_4ft = False    # boolean because it is a true or false data type
features_set = {220, 'seats', 25.8}   # set because it is unordered and has unique elemrnts
price_dictionary = {'Lamborghini': 540_000, 'Corvette': 74_000}   # dictionary because its a collection of key-value pairs
features_tuple = (220, 'seats', 25.8)   # tuple because it uses parentheses, unlike sets, which use curly brackets. tuples like this cannot be changed and are in a fixed order
range_seats = range(2,5,2)    # this code will cause a sequence of numbers to appear.
features_list = [True, 220, 'seats', 25.8]   # list because it contains different data types. this data is in a fixed order
cars_purchase = None   # this tells that no cars have been purchased
#variable names should be in lower-case
AGE = 9
age = 9   # - these are not treated as the same variables
#use descriptive names for variables
house_price = 450,000
hp = 450,000   # - house_price will have its purpose communicated more easily than "hp"

```
